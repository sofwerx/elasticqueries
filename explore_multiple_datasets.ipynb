```python
import glob, os

import pandas as pd
import numpy as np
```

```python
npy_files = glob.glob("data/raw/*.npy")
```

```python
domoticz_npy = [f for f in npy_files if 'domoticz' in f and 'colnames' not in f]
gamma_npy = [f for f in npy_files if 'gamma' in f and 'colnames' not in f]
auditbeat_npy  = [f for f in npy_files if 'auditbeat' in f and 'colnames' not in f]
```

```python
len(domoticz_npy), len(gamma_npy), len(auditbeat_npy)
```

```python
dom_colnames = np.load('data/raw/domoticz_df_colnames.npy')
gamma_colnames = np.load('data/raw/gamma_df_colnames.npy')
auditbeat_colnames = np.load('data/raw/auditbeat_df_colnames.npy')
```

```python
# Read data from npy files and store in pandas dataframes
dom_df = pd.DataFrame()
gamma_df = pd.DataFrame()
auditbeat_df = pd.DataFrame()

for file in domoticz_npy:
    dom_df = pd.concat([pd.DataFrame(np.load(file)), dom_df])
for file in gamma_npy:
    gamma_df = pd.concat([pd.DataFrame(np.load(file)), gamma_df])
for file in auditbeat_npy:
    auditbeat_df = pd.concat([pd.DataFrame(np.load(file)), auditbeat_df])
    
    
dom_df.columns = list(dom_colnames)
gamma_df.columns = list(gamma_colnames)
auditbeat_df.columns = list(auditbeat_colnames)
```
